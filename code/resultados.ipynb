# Visualizacion resultados

```python
import pandas as pd
import numpy as np
```

```python
interpolacion = pd.read_csv("../interpolation/_dataframe_resultados.csv", header=0, index_col=0)
knn = pd.read_csv("../knn/_dataframe_final_resultados.csv", header=0, index_col=0)
mean = pd.read_csv("../mean/_dataframe_resultados.csv", header=0, index_col=0)
median = pd.read_csv("../median/_dataframe_resultados.csv", header=0, index_col=0)
rf = pd.read_csv("../rf/_dataframe_final_resultados.csv", header=0, index_col=0)
svm = pd.read_csv("../svm/_dataframe_final_resultados.csv", header=0, index_col=0)
```

```python
mix_RMSE = pd.concat([mean['Dataset'], mean['RMSE'], median['RMSE'], interpolacion['RMSE'], knn['RMSE'], svm['RMSE'], rf['RMSE']], axis=1, ignore_index=True)
mix_RMSE = mix_RMSE.set_axis(['Datasets', 'Media', 'Mediana', 'Interpolacion', 'KNN', 'SVM', 'RF'], axis=1)
mix_MAE = pd.concat([mean['Dataset'], mean['MAE'], median['MAE'], interpolacion['MAE'], knn['MAE'], svm['MAE'], rf['MAE']], axis=1, ignore_index=True)
mix_MAE = mix_MAE.set_axis(['Datasets', 'Media', 'Mediana', 'Interpolacion', 'KNN', 'SVM', 'RF'], axis=1)
mix_BIAS = pd.concat([mean['Dataset'], mean['BIAS'], median['BIAS'], interpolacion['BIAS'], knn['BIAS'], svm['BIAS'], rf['BIAS']], axis=1, ignore_index=True)
mix_BIAS = mix_BIAS.set_axis(['Datasets', 'Media', 'Mediana', 'Interpolacion', 'KNN', 'SVM', 'RF'], axis=1)
```
